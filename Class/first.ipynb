```python
import matplotlib.pyplot as plt

# small sample data (uses pandas imported in earlier cell)
df_small = pd.DataFrame({
    'category': ['A', 'B', 'C', 'D', 'E'],
    'value': [5, 3, 9, 2, 7]
})

# Bar plot
plt.figure(figsize=(6, 4))
plt.bar(df_small['category'], df_small['value'], color='tab:blue')
plt.title('Sample Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Line plot
plt.figure(figsize=(6, 4))
plt.plot(df_small['category'], df_small['value'], marker='o', linestyle='-', color='tab:orange')
plt.title('Sample Line Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(alpha=0.4)
plt.tight_layout()
plt.show()
```
