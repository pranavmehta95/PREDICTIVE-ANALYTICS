```python
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
from sklearn.linear_model import LinearRegression
data=pd.read_csv('/Users/pranavkumar/Documents/GitHub/PREDICTIVE-ANALYTICS/Class/Position_Salaries.csv')

```

```python

# Define X and y
x = data[['Level']].values
y = data['Salary'].values

poly = PolynomialFeatures(degree=4)
x_poly = poly.fit_transform(x)

model_poly = LinearRegression()
model_poly.fit(x_poly, y)

y_pred = model_poly.predict(x_poly)

print("Predicted salaries:")
print(y_pred)

result = pd.DataFrame({'Level': x.flatten(), 'Actual Salary': y, 'Predicted Salary': y_pred})
print(result)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color='red')
plt.plot(x,y_pred,color='blue')
```
